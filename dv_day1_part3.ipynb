```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
telco= pd.read_csv('D:/DV_Term2/Telco-Customer-Churn.csv')
```

```python
telecom
```

```python
telco.PaymentMethod.unique()
```

```python
#.size will give the count of each
pmt_method= telco.groupby(['PaymentMethod']).size()
pmt_method= pd.DataFrame(pmt_method)
pmt_method= pmt_method.reset_index()
#the column name was 0 earlier so it is being changed to Count
pmt_method= pmt_method.rename(columns= {0:'Count'})
```

```python
pmt_method
```

```python
#shift tab will show the signature, .1f will give 1 digit after point start angle will 
plt.figure(figsize=(9,9))
plt.pie(pmt_method['Count'], labels= pmt_method['PaymentMethod'],autopct="%0.1f%%",startangle=90, shadow= True, explode= (0,0,0.1,0))
plt.axis('equal')
plt.show()
```

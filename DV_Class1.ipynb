```python
#pyplot is a module under matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#prints 10 numbers between 0 and 9 with equal linespace
x= np.linspace(0,9,10)
```

```python
x
```

```python
y= x**2
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("x-y plot")
plt.show()
#here plot.show will remove the area above graph
```

```python
#third parameter is for 1st graph
#subplot divides the grid into 1 rows and 2 columns here  bcz in 1st parameter we have given 1 and in 2nd parameter we have given
#2, 1st parameter---- rows, 2nd parameter ----- columns
plt.subplot(1,2,1)
##divides the point as '*', g represents color and - represents line type
plt.plot(x,y,'-*g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x-y plot')

plt.subplot(1,2,2)
#divides the point as '-'
plt.plot(y,x,'r--')
plt.xlabel('y')
plt.ylabel('x')
plt.title('y-x plot')
```

```python
#Press shift + tab to get the signature
plt.subplot()
```

```python
fig= plt.figure()
#3rd and 4th parameters are changing the size of the graph
#here 1st nd 2nd parameters are changing the position of the origin, compare the below ploted graphs to notice the change
axis1= fig.add_axes([0.1,0.1,0.9,0.9])
axis2= fig.add_axes([0.2,0.2,0.6,0.6])
axis1.plot(x,y)
axis2.plot(x,y)
plt.show()
```

```python
fig= plt.figure()
axes3= fig.add_axes([0.1,0.1,0.9,0.9])
axes3.plot(x,y)
axes3.set_title('x-y plot')
axes3.set_xlabel('x')
axes3.set_ylabel('y')

plt.show()
```

```python
fig, axes5= plt.subplots(nrows=1, ncols= 2)
axes5[0].plot(x,y)
axes5[1].plot(y,x)
axes5[0].set_title('x-y plot')
axes5[1].set_title('y-x plot')
axes5[0].set_xlabel('x')
axes5[0].set_ylabel('y')
axes5[1].set_xlabel('x')
axes5[1].set_ylabel('y')
#this will save the figure in the 
plt.savefig("2plots.png")
plt.show()
```

```python
fig, axes= plt.subplots()
plt.plot(x,y,'m--', label='x-y')
plt.plot(y,x,'g-*', label='y-x')
#loc=5 will change the location of the legend
plt.legend(loc=5)
plt.show()
```

```python
fig,axes= plt.subplots()
#marker will change the shape of the dots
plt.plot(x,y, marker='p', color='chocolate', linewidth=3)
```
